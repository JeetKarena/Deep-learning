```python
import tensorflow as tf
```

```python
print(tf.abs(tf.constant([-1,0,-1])))
```

```python
print(tf.square(tf.constant([4,2])))
```

```python
nums_list_one=tf.constant([4,2,4])
nums_list_two=tf.constant([4,2,4])

print(tf.add(nums_list_one,nums_list_two))
print(tf.subtract(nums_list_one,nums_list_two))
print(tf.divide(nums_list_one,nums_list_two))
print(tf.multiply(nums_list_one,nums_list_two))
```

```python
print(nums_list_two)
print(tf.math.divide_no_nan(nums_list_one,nums_list_two))
```

```python
row_d_tensor=tf.constant([[1,2,3,4]])
col_d_tensor=tf.constant([[1],[2],[3],[4]])
print(row_d_tensor.shape)
print(col_d_tensor.shape)
print(tf.math.multiply(row_d_tensor,col_d_tensor))
```

```python
row_one_tensor=tf.constant([-5,-2,0,2.0,5])
row_two_tensor=tf.constant([0,3,-5,0,8])
```

```python
print(tf.math.maximum(row_one_tensor,row_two_tensor))
```

```python
print(tf.math.minimum(row_one_tensor,row_two_tensor))
```

```python
print(tf.math.equal(row_one_tensor,row_two_tensor))
```

```python
print(tf.math.sigmoid(row_one_tensor))
```
